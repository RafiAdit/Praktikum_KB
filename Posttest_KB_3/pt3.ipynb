Pertama import library yang ingin digunakan dan memanggil file csvnya

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

rafi = pd.read_csv('BodyPerformance.csv')
```

#

Membuat Grafik batang, pertama kiat buat variabel kelas, agar mempermudah kita untuk membuat grafik batang berikut.

```python
kelas = rafi['class'].value_counts()

kelas.plot.bar()
plt.title('Data Performa Badan')
plt.xlabel('Kelas')
plt.ylabel('Total Manusia')
plt.show()
```

```python
dari grafik batang diatas kita mendapatkan informasi bahwa, kelas dari orang orang yg mengikuti pengujian ini sama rata
```

#


membuat histogram untuk membandingkan umur

```python
plt.hist(rafi['age'], bins = 15)
plt.title('Data Total Umur')
plt.ylabel('Total Manusia')
plt.xlabel('Umur')

plt.show()
```

#

membuat piechart, pertama kita buat variabel jenis agar mempermudah untuk membuat piechart

```python
jenis = rafi.gender.value_counts()

plt.pie(jenis, labels=jenis.index, autopct="%1.1f%%", startangle=70)
plt.title('Jumlah Data Jenis Kelamin')
plt.show()
```

dari piechart diatas kita dapat informasi yang mengikuti pengujian performa badan yaitu 63,2% di ikutin oleh laki laki dan sisanya ialah perempuan

#

```python
buat variabel heatmap_data dan correlation_matrix agar mempermudah untuk membuat heatmap
```

```python
heatmap_data = rafi[['age', 'height_cm', 'weight_kg', 'body fat_%', 'diastolic', 'systolic', 'gripForce', 'sit and bend forward_cm', 'sit-ups counts', 'broad jump_cm']]

correlation_matrix = heatmap_data.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, square=True)
plt.title('Correlation Heatmap')
plt.show()
```

warna lebih gelap menunjukkan korelasi yang lebih kuat (positif atau negatif), sedangkan warna lebih terang menunjukkan korelasi yang lebih lemah atau tidak ada korelasi.
