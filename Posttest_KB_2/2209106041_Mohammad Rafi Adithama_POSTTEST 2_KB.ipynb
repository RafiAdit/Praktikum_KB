# 3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka

```python
data = {}
data["Jumlah Record"] = rafi.shape[0]
data["Jumlah Attribute"] = rafi.shape[1]
data["Nama Attribute"] = rafi.columns.tolist()
data["Jumlah Attribute Dengan Tipe Angka"] = rafi.select_dtypes(include=["int64", "float64"]).shape[1]

Keys = list(data.keys())
table = prtl.PrettyTable()
table.add_column(Keys[0], [data["Jumlah Record"]])
table.add_column(Keys[1], [data["Jumlah Attribute"]])
table.add_column(Keys[2], ['\n'.join(data['Nama Attribute'])])
table.add_column(Keys[3], [data["Jumlah Attribute Dengan Tipe Angka"]])
print(table)
```
